```python
#from pandas import DataFrame as
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import tkinter as tk
from fpdf import FPDF


data = {'Date':['2020.04.21', '2020.04.21', '2020.04.21', '2020.04.21'], 'Price':[500, 550, 600, 625]}
prices = pd.DataFrame(data)


#*********************************this is the part for GUI Window******************************#

root= tk.Tk()
 
canvas1 = tk.Canvas(root, width = 100, height = 100)
canvas1.pack()
canvas2 = tk.Canvas(root, width = 100, height = 100)
canvas2.pack()
#**********************************this is the part for GUI window***************************#
price = 500
day_change = 25
week_change = 100

#here we define export function to export images to pdf file
def export():

    with PdfPages('zayn.pdf') as export_pdf: #here mentioned the file name you want to create for saving matplotlib figures
        
        plt.plot(prices['Price'])#here you will plot your matplotlib figures and after that use export_pdf.savefig() function
        export_pdf.savefig()     #to save images in memory so after that you can export to pdf
        #plt.close()
        
        
        plt.plot(prices['Price'])
        export_pdf.savefig()
        #plt.close()
        
        plt.plot(prices['Price'])
        export_pdf.savefig()
        plt.close()
        
        
#This is the function for printing statement in pdf but remember fpdf doesn't allow you to append images so in case you want to
#append images in this file you have to create a merge function so that you can merge two pdf files and create new one
def print_statement():
    '''Print successfully to pdf'''
    pdf = FPDF('P', 'mm', 'A4')
    pdf.add_page('P')
    pdf.set_font('Times', 'B', 14)
   
    pdf.multi_cell(0,5,('welcome to weekly updates:'))  
    pdf.ln()
    pdf.multi_cell(0, 5, ('Price today is: %s' % price ))
    pdf.ln()
    pdf.multi_cell(0, 5, ('Day Change: %s' % day_change +'/Week Change: %s' % week_change))
    pdf.ln()
    pdf.multi_cell(0, 5, ('Many thanks for reading'))
    pdf.ln()
    pdf.output('k.pdf','F')
    
#*************This is for the GUI button area*********************#
button1 = tk.Button (root, text='Export PDF',command=export, bg='brown', fg='white')
button2 = tk.Button (root, text='export output',command=print_statement, bg='brown', fg='white')
canvas1.create_window(50,50,window=button1)
canvas2.create_window(50, 50, window=button2)
    
    
root.mainloop()
```
